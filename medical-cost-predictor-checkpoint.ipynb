```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.ensemble import AdaBoostRegressor,RandomForestRegressor,GradientBoostingRegressor

def rmse_calc(test , pred):
    rmse = mean_squared_error(test,pred, squared = False )
    return rmse
```

## Variable Transformation

```python
data['sex'] = data['sex'].apply(lambda x: 1 if x == 'male' else 0)
data['smoker'] = data['smoker'].apply(lambda x: 1 if x == 'yes' else 0)
data['charges'] = np.log(data['charges'])
X = data.drop(['charges','overweight','region'] , axis = 1)
y = data['charges']
region = pd.get_dummies(data.region)
X = pd.concat([X,region],axis = 1)
x_train , x_test , y_train , y_test = train_test_split(X,y, test_size = 0.2 , random_state =123)
```

## Model Building

#### 1. Linear Regression


```python
linreg = LinearRegression(n_jobs = -1)
linreg.fit(x_train,y_train)
lin_pred = linreg.predict(x_test)
linear_rmse = rmse_calc(y_test , lin_pred)
lin_a_p = np.exp(lin_pred)
test = np.exp(y_test)
print('train' , linreg.score(x_train , y_train))
print('test' , linreg.score(x_test , y_test))

fig , ax = plt.subplots(2,1,figsize = (15,6))
ax[0].plot(lin_a_p , color = 'red')
ax[1].plot(test.tolist() , color = 'green')
ax[0].set_xlabel('predicted')
ax[1].set_xlabel('actual')
print('RMSE : ', linear_rmse)
```

## Ensemble Techniques

#### 1.Random Forest

```python
rfc = RandomForestRegressor(n_estimators = 500 , max_features = 6 , max_depth = 10 , min_samples_leaf = 6)
rfc.fit(x_train,y_train)
print('train' , rfc.score(x_train , y_train))
print('test' , rfc.score(x_test , y_test))

rfc_pred = rfc.predict(x_test)
rfc_rmse = rmse_calc(y_test , rfc_pred)

rfc_p = np.exp(rfc_pred)
rfc_test = np.exp(y_test)

fig , ax = plt.subplots(2,1,figsize = (15,6))
ax[0].plot(rfc_p , color = 'red')
ax[1].plot(rfc_test.tolist() , color = 'green')
ax[0].set_xlabel('predicted')
ax[1].set_xlabel('actual')
print('RMSE : ', rfc_rmse)
```

#### 2.AdaBoost

```python
abr = AdaBoostRegressor(learning_rate = 0.1 )
abr.fit(x_train,y_train)
print('train' , abr.score(x_train , y_train))
print('test' , abr.score(x_test , y_test))
abr_pred = abr.predict(x_test)
abr_rmse = rmse_calc(y_test , abr_pred)

abr_pred = abr.predict(x_test)
abr_p = np.exp(abr_pred)
abr_test = np.exp(y_test)

fig , ax = plt.subplots(2,1,figsize = (15,6))
ax[0].plot(abr_p , color = 'red')
ax[1].plot(abr_test.tolist() , color = 'green')
ax[0].set_xlabel('predicted')
ax[1].set_xlabel('actual')
print('RMSE : ', abr_rmse)
```

#### 3.GradientBoost

```python
gbr = GradientBoostingRegressor()
gbr.fit(x_train,y_train)
print('train' , gbr.score(x_train , y_train))
print('test' , gbr.score(x_test , y_test))

gbr_pred = gbr.predict(x_test)
gbr_rmse = rmse_calc(y_test , gbr_pred)
gbr_p = np.exp(gbr_pred)
gbr_test = np.exp(y_test)

fig , ax = plt.subplots(2,1,figsize = (15,6))
ax[0].plot(gbr_p , color = 'red')
ax[1].plot(gbr_test.tolist() , color = 'green')
ax[0].set_xlabel('predicted')
ax[1].set_xlabel('actual')
print('RMSE : ', gbr_rmse)
```

#### 4.XGBoost

```python
from xgboost.sklearn import XGBRegressor

xgb = XGBRegressor(gamma = 0.5)
xgb.fit(x_train,y_train)
print('train' , xgb.score(x_train , y_train))
print('test' , xgb.score(x_test , y_test))

xgb_pred = xgb.predict(x_test)
xgb_rmse = rmse_calc(y_test , xgb_pred)
xgb_p = np.exp(xgb_pred)
xgb_test = np.exp(y_test)

fig , ax = plt.subplots(2,1,figsize = (15,6))
ax[0].plot(xgb_p , color = 'red')
ax[1].plot(xgb_test.tolist() , color = 'green')
ax[0].set_xlabel('predicted')
ax[1].set_xlabel('actual')
print('RMSE : ', xgb_rmse)
```
